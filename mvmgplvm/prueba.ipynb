```python
import numpy as np
import matplotlib.pyplot as plt
gpmethod='vmgplvm'
surface1='cat0'
surface2='cat10'
saveruta='results/'+gpmethod+'/geodesicdistance-nonrigid/'+surface1+'-'+surface2+'.txt'
geodist=np.loadtxt(saveruta)
```

```python
step=0.005
thresh = np.arange(0,0.25+step,step)
tamth=thresh.shape[0]

curve=[]
for i in range(tamth):
    curve.append(100*np.sum(geodist<=thresh[i])/geodist.shape[0])
plt.plot(thresh,curve)
```
